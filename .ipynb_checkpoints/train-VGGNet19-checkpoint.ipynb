```python
def show_ingroup(im):   # Display a random image tensor
  c=0
  if im.shape[-1]<12:
    for _ in range(im.shape[-1]//3):
      x = im[:, :, c:c+3]
      plt.imshow(x)
      plt.title(str(x.shape)+' | '+str(x.dtype), fontsize=15)
      c+=3
      plt.show()
  else:
    for _ in range(im.shape[-1]//12):
      fig = plt.figure(figsize=(20, 20))
      ax1 = fig.add_subplot(1,4,1)
      x=im[:, :, c:c+3]
      ax1.imshow(x)
      ax1.set_title(str(x.shape)+' | '+str(x.dtype),fontsize=15)
      c+=3
      ax2 = fig.add_subplot(1,4,2)
      x=im[:, :, c:c+3]
      ax2.imshow(x)
      ax2.set_title(str(x.shape)+' | '+str(x.dtype),fontsize=15)
      c+=3
      ax3 = fig.add_subplot(1,4,3)
      x=im[:, :, c:c+3]
      ax3.imshow(x)
      ax3.set_title(str(x.shape)+' | '+str(x.dtype),fontsize=15)
      c+=3
      ax4 = fig.add_subplot(1,4,4)
      x=im[:, :, c:c+3]
      ax4.imshow(x)
      ax4.set_title(str(x.shape)+' | '+str(x.dtype),fontsize=15)
      c+=3
    plt.show()

def to_DownscaledTiles(image,DownscaleTimes):
  k=[]
  c=0
  for i in range(image.shape[-1]//3):
    x=pyramid_reduce(image[:,:,c:c+3], downscale=2, sigma=None, order=1, mode='reflect', cval=0, multichannel=True)
    x *= 255 
    x = x.astype(np.uint8)
    x=np.concatenate((x,x), axis=0)
    x=np.concatenate((x,x), axis=1)
    k.append(x)
    c+=3
  if DownscaleTimes==2:
    c=0  
    for i in range(image.shape[-1]//3):
      x=pyramid_reduce(image[:,:,c:c+3], downscale=4, sigma=None, order=1, mode='reflect', cval=0, multichannel=True)
      x *= 255 
      x = x.astype(np.uint8)
      x=np.concatenate((x,x,x,x), axis=0)
      x=np.concatenate((x,x,x,x), axis=1)
      k.append(x)
      c+=3
  for i in k:
    image=np.concatenate((image, i), axis=-1)
  return image
  
def to_RGB(im):
    w, h = im.shape
    ret = np.empty((w, h, 3), dtype=np.uint8)
    #ret[:, :, 0] = im
    #ret[:, :, 1] = im
    #ret[:, :, 2] = im
    ret[:, :, :] = im[:, :, np.newaxis]
    return ret
  
def process_images(LowerIndex,UpperIndex,PatchSize,overlap,overlap_PatchCount,DownscaleTiled,DownscaleTimes):
    """
    Returns two list: 
        X is an list of images
        Y is an list of labels
    """ 

    X = [] #list to store image/patch data
    Y = [] #list to store corresponding class label
    #div = int(np.sqrt(divisor))
    FnameCount=0
    PatchCount=0
    for G in FnameList[LowerIndex:UpperIndex]:
      classZero = fnmatch.filter(G,'*neg.tif') #list to store fname ending ...neg.tif
      classOne = fnmatch.filter(G,'*pos.tif')  #list to store fname ending ...pos.tif
      for j in range(len(G)):
        print(FnameCount+1,':',G[j])
        z = np.array(io.imread(G[j]))
        if len(z.shape)==2:
          z = to_RGB(z)
        if j>0:
          image=np.concatenate((image, z), axis=-1)
        else:
          image = z
      print(image.shape)
        
      if DownscaleTiled:
        image = to_DownscaledTiles(image,DownscaleTimes)  
        print(image.shape)
        
      show_ingroup(image)
      
      __builtins__.W = PatchSize
      __builtins__.H = PatchSize
      __builtins__.C = image.shape[-1]

      ScrapCount=0
      if overlap == False:
        PatchList = [image[x:x+W,y:y+H,:] for x in range(0,image.shape[0],W) for y in range(0,image.shape[1],H)]
      else:
        PatchList=[]
        idxs = []
        target_height=target_width = PatchSize
        for i in range(0, overlap_PatchCount):
          start_row_idx = random.randint(0, image.shape[1]-target_height-1)
          start_col_idx = random.randint(0, image.shape[0]-target_width-1)
          if (start_row_idx,start_col_idx) not in idxs:
            p = image[start_row_idx:(start_row_idx+target_height), start_col_idx:(start_col_idx+target_width), :]
            if (np.array(p)).shape == (W,H,C):
              PatchList.append(p)
              idxs.append((start_row_idx, start_col_idx))
            else:
              ScrapCount+=1
      
      for Patch in PatchList:
        if (np.array(Patch)).shape != (W,H,C):
          ScrapCount+=1
          continue
        X.append(Patch)
        if G == classZero:
          Y.append(0)
        elif G == classOne:
          Y.append(1)
        else:
          return
      
      print('Each  patch shape from image',FnameCount,':', X[0].shape)
      PrePatchCount=PatchCount
      PatchCount += len(PatchList)-ScrapCount
      print('Total patch count upto image',FnameCount,':',len(PatchList),'-',ScrapCount,'=',len(PatchList)-ScrapCount,'+',PrePatchCount,'=',PatchCount,'\n')

      FnameCount+=1
      if overlap:
        del idxs
      del PatchList

    return X,Y
```

```python
#from efficientnet import EfficientNetB3
from keras_efficientnets import EfficientNetB3
from keras.applications.vgg19 import VGG19
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten
from keras.layers import Conv2D, MaxPooling2D
from keras.layers import Input
from keras.models import Model
from keras.optimizers import Adam
from keras import regularizers

# conv_base = EfficientNetB3(weights='imagenet',
#                            input_shape=(300,300,3),
#                            include_top=True,
#                            pooling=None)
# conv_base.summary()



# conv_base = keras.applications.vgg19.VGG19(input_shape=(W,H,3), 
#                                            include_top=False, 
#                                            weights='imagenet', 
#                                            input_tensor=None, 
#                                            pooling=None)
# conv_base.summary()



def input_block_model(w, h, c):
    img_input = Input(shape=(w,h,c))
    x = Conv2D(9, (3,3), padding= 'same')(img_input)
    x = Conv2D(6, (3,3), padding= 'same')(x)
    x = Conv2D(3, (3,3), padding= 'same')(x)
    m = Model(img_input, x)
    return m
input_model = input_block_model(W,H,C)
pretrained_model = VGG19(weights = "imagenet", include_top=False, pooling=None)
#pretrained_model = EfficientNetB1(weights = "imagenet", include_top=False, pooling=None)
conv_base = Model(input_model.input, pretrained_model(input_model.output), name='Combined_conv_base')
conv_base.summary()




model = Sequential()
model.add(conv_base)
model.add(Flatten())
model.add(Dense(128, activation='relu', kernel_regularizer=regularizers.l2(0.001)))
model.add(Dropout(0.5))
model.add(Dense(128, activation='relu', kernel_regularizer=regularizers.l2(0.001)))
model.add(Dropout(0.5))
model.add(Dense(num_classes, activation='sigmoid'))

model.name = Model_name
model.summary()
```

```python
import json,codecs
import numpy as np

def save_history(path,history):
    new_hist = {}
    for key in list(history.history.keys()):
        if type(history.history[key]) == np.ndarray:
            new_hist[key] == history.history[key].tolist()
        elif type(history.history[key]) == list:
           if  type(history.history[key][0]) == np.float64:
               new_hist[key] = list(map(float, history.history[key]))

    print(new_hist)
    with codecs.open(path, 'w', encoding='utf-8') as f:
        json.dump(new_hist, f, separators=(',', ':'), sort_keys=True, indent=4) 

def load_history(path):
    with codecs.open(path, 'r', encoding='utf-8') as f:
        n = json.loads(f.read())
    return n
```
